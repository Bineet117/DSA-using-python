## generic tree

```python
class RootNode():

    def __init__(self, data):
        self.data = data
        self.children = []


root = RootNode(1)
children1 = RootNode(2)
children2 = RootNode(3)
children3 = RootNode(4)
children4 = RootNode(5)

root.children.append(children1)
root.children.append(children2)
root.children.append(children3)
root.children.append(children4)


# printing tree 
def print_tree(root):
    print(f"{root.data}:", end="") 
    for eachchild in root.children:
        print(eachchild.data, end=",")

    print()

    for children in root.children:
        print_tree(children)

print_tree(root)
```

```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = []

    def print_tree(self, root_node):
        if root_node == None:
            return

        print(f"{root_node.data} : ", end=" ")

        for child in root_node.children:
            print(child.data , end=",")

        print()

        for child in root_node.children:
            self.print_tree(child)

      

root = TreeNode(1)

child1 = TreeNode(20)
child2 = TreeNode(30)
child3 = TreeNode(40)

root.children.append(child1)
root.children.append(child2)
root.children.append(child3)

root.print_tree(root)
```
